```python
# === Part 4: California Housing Regression (clean stable version) ===
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.inspection import permutation_importance

from IPython.display import display
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset as DataFrame with feature names
cal = fetch_california_housing(as_frame=True)
X = cal.data.copy()
y = cal.target.copy()
feature_names = X.columns.tolist()

# Train/test split (fixed random_state for reproducibility)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42
)

print("X_train shape:", X_train.shape, " | X_test shape:", X_test.shape)
print("Features:", feature_names)

```

```python
def report_regression(pipeline, name: str):
    """Fit a regression pipeline, evaluate metrics, and produce diagnostic plots."""
    pipeline.fit(X_train, y_train)
    pred = pipeline.predict(X_test)

    r2 = r2_score(y_test, pred)
    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))

    print(f"[{name}] R^2={r2:.4f} | MAE={mae:.4f} | RMSE={rmse:.4f}")

    # Predicted vs Actual scatter plot
    plt.figure()
    plt.scatter(y_test, pred, alpha=0.35)
    lims = [min(y_test.min(), pred.min()), max(y_test.max(), pred.max())]
    plt.plot(lims, lims, linestyle="--", color="red")  # y=x line
    plt.xlabel("Actual Median House Value")
    plt.ylabel("Predicted")
    plt.title(f"{name}: Predicted vs Actual")
    plt.show()

    # Residual histogram
    resid = y_test - pred
    plt.figure()
    plt.hist(resid, bins=30, edgecolor="black")
    plt.xlabel("Residual")
    plt.ylabel("Frequency")
    plt.title(f"{name}: Residuals Histogram")
    plt.show()

    return pipeline, {"R2": r2, "MAE": mae, "RMSE": rmse}

# Define regression models
pipe_linear = Pipeline([("reg", LinearRegression())])

pipe_ridge = Pipeline([
    ("scale", StandardScaler()),
    ("reg", Ridge(alpha=0.7))  # no random_state for Ridge in some versions
])

pipe_lasso = Pipeline([
    ("scale", StandardScaler()),
    ("reg", Lasso(alpha=0.001, max_iter=50000))
])

results = {}
for name, model in [
    ("Linear", pipe_linear),
    ("Ridge(α=0.7)", pipe_ridge),
    ("Lasso(α=0.001)", pipe_lasso),
]:
    fitted, scores = report_regression(model, name)
    results[name] = scores

# Optional: coefficient table for linear models
def coef_table(fitted_pipeline, label):
    """Print top coefficients (absolute value) for linear-type models."""
    reg = fitted_pipeline.named_steps.get("reg", None)
    if reg is None or not hasattr(reg, "coef_"):
        return
    coefs = np.asarray(reg.coef_)
    order = np.argsort(-np.abs(coefs))
    print(f"\nTop coefficients for {label}:")
    for idx in order[:10]:
        print(f"  {feature_names[idx]:<20} | {coefs[idx]: .4f}")

coef_table(pipe_linear, "Linear")
coef_table(pipe_ridge, "Ridge")
coef_table(pipe_lasso, "Lasso")
```

```python
# Permutation importance on the best model by R^2 (robust to feature count)
best_name = max(results, key=lambda k: results[k]["R2"])
print("\nBest model based on R^2:", best_name)

best_pipe = {
    "Linear": pipe_linear,
    "Ridge(α=0.7)": pipe_ridge,
    "Lasso(α=0.001)": pipe_lasso
}[best_name]

best_pipe.fit(X_train, y_train)

pi = permutation_importance(best_pipe, X_test, y_test, n_repeats=10, random_state=42)

n_feats = len(feature_names)
top_k = min(10, n_feats)                 # prevent shape mismatch when features < 10
sorted_idx = np.argsort(-pi.importances_mean)[:top_k]

vals = pi.importances_mean[sorted_idx]   # length = top_k
labels = [feature_names[i] for i in sorted_idx]

# plot from small to large for a nicer horizontal barh layout
order = np.argsort(vals)
vals_plot = vals[order]
labels_plot = [labels[i] for i in order]

plt.figure(figsize=(6, 4))
ypos = np.arange(top_k)
plt.barh(ypos, vals_plot)
plt.yticks(ypos, labels_plot)
plt.xlabel("Permutation Importance (mean decrease in score)")
plt.title(f"{best_name}: Top-{top_k} Feature Importance")
plt.tight_layout()
plt.show()
```
